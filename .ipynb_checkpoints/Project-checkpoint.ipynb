```python
# Split the data into features (X) and target (y)
X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

```python
# Scale the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Define the machine learning models
models = [
    LogisticRegression(),
    KNeighborsClassifier(),
    SVC(),
    MLPClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    LGBMClassifier()
]
```

```python
# Define the hyperparameter tuning space for each model
param_grids = [
    {'penalty': ['l1', 'l2'], 'C': [0.1, 1, 10]},
    {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']},
    {'kernel': ['linear', 'poly', 'rbf'], 'C': [0.1, 1, 10]},
    {'hidden_layer_sizes': [(10,), (20,), (30,)], 'activation': ['relu', 'tanh']},
    {'max_depth': [3, 5, 7], 'min_samples_split': [2, 5, 10]},
    {'n_estimators': [10, 50, 100], 'max_depth': [3, 5, 7]},
    {'learning_rate': [0.1, 0.5, 1], 'n_estimators': [10, 50, 100]},
    {'num_leaves': [31, 62, 127], 'learning_rate': [0.1, 0.5, 1]}
]
```

```python
# Perform hyperparameter tuning using GridSearchCV
results = []
for model, param_grid in zip(models, param_grids):
    grid = GridSearchCV(model, param_grid, cv=5, scoring='f1_macro')
    grid.fit(X_train, y_train)
    results.append((model.__class__.__name__, grid.best_score_, grid.best_params_))
```

```python
# Print the results
print("Model\tBest Score\tBest Parameters")
for result in results:
    print(f"{result[0]}\t{result[1]:.3f}\t{result[2]}")
```

```python
# Train the best model on the entire training set
best_model = GradientBoostingClassifier(**results[6][2])
best_model.fit(X_train, y_train)
```

```python
# Evaluate the best model on the testing set
y_pred = best_model.predict(X_test)
y_pred_proba = best_model.predict_proba(X_test)[:, 1]

print("Best Model: GradientBoostingClassifier")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_pred_proba))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Define models and parameter grids
models = [GradientBoostingClassifier(), RandomForestClassifier(), SVC(probability=True)]
param_grids = [
    {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1, 0.2]},
    {'n_estimators': [100, 200], 'max_features': ['auto', 'sqrt', 'log2']},
    {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
]

results = []
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for model, param_grid in zip(models, param_grids):
    grid = GridSearchCV(model, param_grid, cv=cv, scoring='f1_macro', n_jobs=-1)
    grid.fit(X_train, y_train)
    results.append((model.__class__.__name__, grid.best_score_, grid.best_params_))

# Print the results
for result in results:
    print(f"Model: {result[0]}")
    print(f"Best F1 Score: {result[1]}")
    print(f"Best Parameters: {result[2]}")
```
